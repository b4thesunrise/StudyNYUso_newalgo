```python
import requests
```

```python
realtimeformat = 'http://mis.nyiso.com/public/csv/realtime/{}realtime_zone_csv.zip'
dayaheadformat = 'http://mis.nyiso.com/public/csv/damlbmp/{}damlbmp_zone_csv.zip'
yearlist = [str(x) for x in range(1999,2021)]
monthlist = [str(x) for x in range(1,13)]
monthlist = ['0' + x if len(x) < 2 else x for x in monthlist]
print(yearlist)
print(monthlist)
```

```python
startmode = False
for year in yearlist:
    print(year)
    for month in monthlist:
        ym = year + month
        if ym == '200808':
            startmode = True
        if not startmode:
            continue
        rt_fileloc = realtimeformat.format(ym + '01')
        rt = requests.get(rt_fileloc, stream = True)
        with open('Realtime/' + ym + '.zip', 'wb') as f:
            for chunk in rt.iter_content(chunk_size = 512):
                if chunk:
                    f.write(chunk)
        da_fileloc = dayaheadformat.format(ym + '01')
        da = requests.get(da_fileloc, stream = True)
        with open('Dayahead/' + ym + '.zip', 'wb') as f:
            for chunk in da.iter_content(chunk_size = 512):
                if chunk:
                    f.write(chunk)
```

```python
import os
import shutil
def unzip(target_path, filepath):
    os.makedirs(target_path)
    shutil.unpack_archive(filepath,target_path)
```

```python
for year in yearlist:
    if year == '1999':
        continue
    print(year)
    for month in monthlist:
        ym = year + month
        target_rt = 'Realtime' + '/' + ym
        file_rt = target_rt + '.zip'
        target_da = 'Dayahead' + '/' + ym
        file_da = target_da + '.zip'
        unzip(target_rt, file_rt)
        unzip(target_da, file_da)
```

```python
for year in yearlist:
    print(year)
    for month in monthlist:
        ym = year + month
        target_rt = 'Realtime' + '/' + ym
        file_rt = target_rt + '.zip'
        target_da = 'Dayahead' + '/' + ym
        file_da = target_da + '.zip'
        os.remove(file_rt)
        os.remove(file_da)
```
