```python
def heatplot(df, vmin):
    fig = plt.figure() #调用figure创建一个绘图对象
    ax = fig.add_subplot(111)
    cax = ax.matshow(df, vmin=vmin, vmax=1)  #绘制热力图，从-1到1
    fig.colorbar(cax)  #将matshow生成热力图设置为颜色渐变条
    ticks = np.arange(0,len(df.index.tolist()),1) 
    ax.set_xticks(ticks)  #生成刻度
    ax.set_yticks(ticks)
    ax.set_xticklabels(df.index.tolist()) #生成x轴标签
    ax.set_yticklabels(df.index.tolist())
    plt.show()
```

```python
def dailyploting(daily_np):
    mean_np = np.mean(daily_np, 0)
    std_np = np.std(daily_np, 0)
    plt.figure()
    for i in range(daily_np.shape[2]):
        plt.subplot(4, 4, i+1)
        plt.plot(mean_np[:,i])
        plt.fill_between(range(0,24), mean_np[:,i] + std_np[:,i], mean_np[:,i] - std_np[:,i], alpha = 0.4)
    plt.show()
```

```python
dailyploting(daily_danp)
```

```python
dailyploting(daily_rtnp)
```

```python
def yearploting(year_np):
    mean_np = np.mean(year_np, 0)
    std_np = np.std(year_np, 0)
    plt.figure()
    for i in range(year_np.shape[2]):
        plt.subplot(4, 4, i+1)
        plt.plot(mean_np[:,i])
        plt.fill_between(range(0,365), mean_np[:,i] + std_np[:,i], mean_np[:,i] - std_np[:,i], alpha = 0.4)
    plt.show()
```

```python
dadf.index = pd.to_datetime(dadf.index)
rtdf.index = pd.to_datetime(rtdf.index)
```

```python
year_dadf = dadf.resample('1D').mean()
year_rtdf = rtdf.resample('1D').mean()
```

```python
from datetime import timedelta
def get_period_np(year_df, days = 365):
    index_list = year_df.index.tolist()
    start = index_list[0]
    td = timedelta(days = days)
    ret_array = np.zeros((0, 365, 15))
    while True:
        print(start)
        try:
            startloc = index_list.index(start)
            endloc = index_list.index(start + td)
            tempdf = year_df.iloc[startloc:endloc].to_numpy().reshape(-1, 365, 15)
            print(tempdf.shape, ret_array.shape)
            ret_array = np.concatenate((ret_array, tempdf), axis = 0)
            start += td
        except Exception as e:
            print(e)
            break
    return ret_array
```

```python
year_danp = get_period_np(year_dadf)
print(year_danp.shape)
```

```python
yearploting(year_danp)
```

```python
year_rtnp = get_period_np(year_rtdf)
yearploting(year_rtnp)
```

#### 对季度取mean和对小时取mean发现小时影响比较小，季度影响可能相对比较大，接下来考虑控制在某个小时的季度影响 / 控制在某个季度的时间周期

```python
dadf['Month'] = (dadf.index.month)
dadf['Month'] = dadf['Month'].astype(int)
dadf['Hour'] = (dadf.index.hour / 2)
dadf['Hour'] = dadf['Hour'].astype(int)
```

```python
rtdf['Month'] = (rtdf.index.month)
rtdf['Month'] = rtdf['Month'].astype(int)
rtdf['Hour'] = (rtdf.index.hour / 2)
rtdf['Hour'] = rtdf['Hour'].astype(int)
```

```python
# 分析某个小时下的月份影响
def analysis_M(df):
    mean_df = df.groupby(['Month']).mean().drop(columns = ['Hour'])
    std_df = df.groupby(['Month']).std().drop(columns = ['Hour'])
    plt.figure()
    for i,name in enumerate(mean_df.columns):
        plt.subplot(4, 4, i+1)
        plt.plot(mean_df[name])
        plt.fill_between(range(0,12), mean_df[name] + std_df[name], mean_df[name] - std_df[name], alpha = 0.4)
    plt.show()
```

```python
for hour in range(0, 12):
    temphour_dadf = dadf[dadf['Hour'] == hour]
    print(hour)
    analysis_M(temphour_dadf)
```

```python
for hour in range(0, 12):
    temphour_rtdf = rtdf[rtdf['Hour'] == hour]
    print(hour)
    analysis_M(temphour_rtdf)
```

```python
# 分析某个小时下的天影响
def analysis_Day(df):
    mean_df = df.groupby(['Quarter']).mean().drop(columns = ['Month', 'Hour'])
    std_df = df.groupby(['Quarter']).std().drop(columns = ['Month', 'Hour'])
    std_df.index = range(0,366)
    mean_df.index = range(0,366)
    plt.figure()
    for i,name in enumerate(mean_df.columns):
        plt.subplot(4, 4, i+1)
        plt.plot(mean_df[name])
        plt.fill_between(range(0,366), mean_df[name] + std_df[name], mean_df[name] - std_df[name], alpha = 0.4)
    plt.show()
```

```python
for hour in range(0, 12):
    temphour_rtdf = rtdf[rtdf['Hour'] == hour]
    print(hour)
    analysis_Day(temphour_rtdf)
```

#### 最后，进行时间序列的变量选择

```python
def construct_regdf(dadf, rtdf, name, number):
    retdf = pd.DataFrame(columns = [name])
    retdf[name] = rtdf[name]
    for i in range(1,number+1):
        retdf['da_' + str(i)] = dadf[name].shift(i)
        retdf['rt_' + str(i)] = rtdf[name].shift(i)
    retdf = retdf.dropna()
    return retdf
```

```python
loc_namelist = dadf.columns.tolist()
loc_namelist = loc_namelist[0:15]
```

```python
test_hours = 72
```

```python
for name in loc_namelist:
    countlist = []
    print(name)
    regdf = construct_regdf(rtdf, dadf, name, test_hours)
    reg1.fit(regdf.drop(columns = [name]), regdf[name])
    for i,coef in enumerate(reg1.coef_):
        if coef != 0:
            countlist.append(i)
    plt.hist(countlist, bins = 142)
    plt.show()
```
