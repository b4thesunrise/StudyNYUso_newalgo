```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
yearlist = [str(x) for x in range(2000,2021)]
monthlist = [str(x) for x in range(1,13)]
monthlist = ['0' + x if len(x) < 2 else x for x in monthlist]
print(yearlist)
print(monthlist)
```

```python
import os
baseda = 'Dayahead/{}'
basert = 'Realtime/{}'
dafilelist = []
rtfilelist = []
for year in yearlist:
    for month in monthlist:
        ym = year + month
        da = baseda.format(ym)
        rt = basert.format(ym)
        if os.path.exists(da):
            for f in os.listdir(da):
                dafilelist.append(f)
            for f in os.listdir(rt):
                rtfilelist.append(f)
```

```python
dafilelist = sorted(dafilelist)
rtfilelist = sorted(rtfilelist)
```

```python
dadays = [x[0:8] for x in dafilelist]
rtdays = [x[0:8] for x in rtfilelist]
```

```python
notindays = []
for day in dadays:
    if day not in rtdays:
        notindays.append(day)
```

```python
for day in rtdays:
    if day not in dadays:
        print(day)
```

```python
adjust_dafilelist = []
for x in dafilelist:
    isin = False
    for y in notindays:
        if y in x:
            print(x)
            isin = True
            break
    if not isin:
        adjust_dafilelist.append(x)
```

```python
import pickle
def savefile(path, obj):
    with open(path,'wb') as f:
        pickle.dump(obj, f)
def loadfile(path):
    with open(path, 'rb') as f:
        return pickle.load(f)
```

```python
savefile('Database/adjust_dafilelist.pkl', adjust_dafilelist)
savefile('Database/dafilelist.pkl', dafilelist)
savefile('Database/rtfilelist.pkl', rtfilelist)
```
